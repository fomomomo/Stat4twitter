```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Scrapping data 

```python
URL = "https://sites.google.com/site/britishmonarchyfamilyhistory/british-monarchs-and-their-prime-ministers"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
all_p = soup.find_all('p')
```

```python
all_b = []
for p in all_p:
    all_b.append(p.b)
```

```python
all_b
```

```python
b
```

```python
b = []
for some_b in all_b:
    if(some_b!=None and some_b.font==None):
        b.append(some_b)
```

```python
Ruler = []
temp = []
for item in b:
    if(item.has_attr('style')):
        if(len(temp)!=0):
            Ruler.append(temp)
            temp = []
            temp.append(item.get_text())
        else:
            temp.append(item.get_text())
    else:
        temp.append(item.get_text())
        
```

```python
Ruler
```

```python
Ruler = Ruler[:10]
```

```python
Ruler[-1] = Ruler[-1][:-3]
```

```python
Ruler
```

```python
final_list = [{'ruler':ruler[0], 'pm':len(ruler)-1} for ruler in Ruler]
```

```python
final_list
```

```python
df = pd.DataFrame(final_list)
```

```python
df
```

```python
data = df.pm
labels = df.ruler

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:10]
explode = [0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.02,0.3]
#create pie chart
plt.pie(data, labels = labels, colors = colors, explode=explode)
plt.legend(df.pm)
plt.show()
```
